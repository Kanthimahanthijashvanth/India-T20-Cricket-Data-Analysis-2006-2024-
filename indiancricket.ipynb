```python
import matplotlib.pyplot as plt

# Filter Australia games
australia_games = df.query("`Team 1` == 'Australia' or `Team 2` == 'Australia'")

# Extract numeric margin
australia_games['Numeric_Margin'] = australia_games['Margin'].str.extract('(\d+)').astype(float)

# Drop NaNs
scatter_data = australia_games.dropna(subset=['Numeric_Margin', 'Year'])

plt.figure(figsize=(10,6))
plt.scatter(scatter_data['Year'], scatter_data['Numeric_Margin'], color='orange', alpha=0.7)

# Add data labels for each point
for _, row in scatter_data.iterrows():
    plt.text(row['Year'], row['Numeric_Margin'] + 0.5,  # slightly above point
             str(int(row['Numeric_Margin'])), fontsize=8, ha='center')

plt.title('Australia Matches: Winning Margin over Years')
plt.xlabel('Year')
plt.ylabel('Winning Margin')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np

# Load dataset (assuming you have it)
df = pd.read_csv("https://raw.githubusercontent.com/badapa1502/Cricket-Data-Analysis-T20/refs/heads/main/Datasets/india_T20_matches_2006-2024.csv")

# Filter matches where Team 1 or Team 2 is India
india_games = df.query("`Team 1` == 'India' or `Team 2` == 'India'")

# Assume 'Captain' column exists in the dataset; if not, you need to add it or get data from elsewhere

# For demonstration, let’s simulate a 'Captain' column (replace this with real data)
# For example, randomly assign captains from a list
import random
captains_list = ['Dhoni', 'Kohli', 'Rohit', 'Ganguly', 'Sehwag']
india_games['Captain'] = india_games.apply(lambda x: random.choice(captains_list), axis=1)

# Group stats by captain (example: count matches, wins, batting avg, wickets etc.)
# Since real batting avg and wickets per captain might not be in the dataset, we’ll do simple counts here:

# Total matches captained
matches_captained = india_games.groupby('Captain').size()

# Wins as captain
wins_captain = india_games[india_games['Winner'] == 'India'].groupby('Captain').size()

# Win percentage
win_percentage = (wins_captain / matches_captained * 100).fillna(0)

# Create a summary DataFrame
captain_stats = pd.DataFrame({
    'Matches': matches_captained,
    'Wins': wins_captain,
    'Win %': win_percentage
}).fillna(0)

print(captain_stats)
```

```python
import matplotlib.pyplot as plt

# Normalize data for radar chart (scale 0-1)
stats = captain_stats[['Matches', 'Win %']].copy()
stats = (stats - stats.min()) / (stats.max() - stats.min())

# Radar plot setup
labels = list(stats.columns)
num_vars = len(labels)

angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]

fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))

for i, row in stats.iterrows():
    values = row.tolist()
    values += values[:1]
    ax.plot(angles, values, label=i)
    ax.fill(angles, values, alpha=0.25)

ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels)
ax.set_yticklabels([])
ax.set_title("Performance Metrics of India Captains (Normalized)")

plt.legend(loc='upper right', bbox_to_anchor=(1.3, 1))
plt.show()
```
